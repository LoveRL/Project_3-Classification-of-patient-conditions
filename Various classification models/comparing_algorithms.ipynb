```python
import os
import glob as g
import numpy as np
import sklearn.metrics as metrics

# Boosting algorithms
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

# The others algorithms
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

os.chdir('C:\\Users\\seo\\Desktop\\Lab\\Projects\\정부) 재활로봇\\임상 data')
#os.chdir('D:\\Personal_Folders\\kanghyeon108')

def ShowResult(test_labels, estimated_labels, idx) :
    
    print(idx,'-Fold\n')
    
    print('[Accuracy Rate]\n')
    print(round(metrics.accuracy_score(test_labels, estimated_labels), 2), end='\n\n')

    print('[Confusion Matrix]\n')
    print(metrics.confusion_matrix(test_labels, estimated_labels), end='\n\n')

    #print('[Classification Report]\n')
    #target_names = ['Class '+str(i) for i in range(8)]
    #print(metrics.classification_report(test_labels, estimated_labels, target_names=target_names))
    
    print('\n============================================================\n')
    
    return

def SplitData(file_name) :
    
    data = np.loadtxt(file_name, delimiter=',', dtype=np.float32)
    
    np.random.shuffle(data)
    
    features, labels = data[:, 0:-1], data[:, -1]
    
    return features, labels

Avg_Accuracy = 0

for ix, train_file, test_file in zip([i for i in range(1,6)], g.glob('train_*.csv'), g.glob('test_*.csv')) :
    
    train_features, train_labels = SplitData(train_file)
    test_features, test_labels = SplitData(test_file)
    
    #model = RandomForestClassifier(class_weight={0:1, 1:2, 2:2, 3:2, 4:1, 5:1, 6:1, 7:1})
    model = KNeighborsClassifier()
    #model = DecisionTreeClassifier(class_weight={0:1, 1:2, 2:2, 3:2, 4:1, 5:1, 6:1, 7:1})
    
    #model = XGBClassifier(objective='multi:softmax', learning_rate=0.5)
    #model = LGBMClassifier(boosting_type='gbdt', objective='multiclass', learning_rate=0.5)

    model = model.fit(train_features, train_labels)
    
    #print(model.n_features_)
    #print(model.feature_importances_)
    
    estimated_labels = model.predict(test_features)
    
    Avg_Accuracy += metrics.accuracy_score(test_labels, estimated_labels)
    ShowResult(test_labels, estimated_labels, ix)
    
    del model
    
print('Avg. Accuracy rate :', round(Avg_Accuracy / 5, 2), end = '\n\n')
```
